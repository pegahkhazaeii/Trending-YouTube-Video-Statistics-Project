## Association Rule Mining

üí° Here, our goal is to find association rules for ***tags*** attribute in the dataset.

üí° To fullfil this aim, we applied ***apriori algorithm***, and we used [this link](http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/) for ***apriori algorithm***.


```python
df.tags
```

```python
# In the first step, we splited tags in each row ‚¨áÔ∏è
```

```python
tags = df.tags.str.replace('"' , '').str.split("|").values #.values converts series to array.
```

```python
# Now in each row, the "tags" attribute is converted to a list of seperated tags.
tags
```

```python
# Here, we count the repetition number of each tag for the sake of determining the "sup" value in apriori algorithm
counted_tags = {}
for i in range(tags.shape[0]):
    one_tag_count = Counter(tags[i])
    for item in one_tag_count.keys():
        if item in counted_tags.keys():
            counted_tags[item] += one_tag_count[item]
        else:
            counted_tags[item] = one_tag_count[item]
```

```python
counted_tags
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
tgs = []
for i in range(20000):
    x = df.tags[i]
    y = x.replace('"', '').split("|")
    tgs.append(y)
tgs 
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(tgs).transform(tgs)
```

```python
df2 = pd.DataFrame(te_ary, columns = te.columns_)
frequent_itemsets = apriori(df2, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric = "confidence", min_threshold = 0.7)
```

## Working on ***publish_time*** and ***trending_date*** attributes




üí° The format of ***"publish_time"*** is as follows: 
   "2017-11-10T17:00:03.000Z" (Year-Month-Day T Hour:Minute:Second.000Z).

üí° Also, the format of ***"trending_data"*** is as follows: 
   "17.14.11" (Year.Day.Month).
   
   
üí° Here, the goal is to seperate ***date*** from ***time*** in ***"publish_time"*** attribute
   and, to alter the format of ***"trending_date"*** to Year.Month.Day.
   
üí° [A Useful Link](https://stackoverflow.com/questions/51462050/how-to-remove-t-and-z-in-time-in-python)

```python
df["seperated_publish_time"] = pd.to_datetime(df["publish_time"], format="%Y-%m-%dT%H:%M:%S.%fZ")
```

```python
df["published_date"] = [d.date() for d in df['seperated_publish_time']]
df["published_time"] = [d.time() for d in df['seperated_publish_time']]
```

```python
df['trending_date'] =pd.to_datetime(df['trending_date'], format='%y.%d.%m').dt.date
```

```python
df["published_trending_difference"] = df["trending_date"] - df["published_date"]
```

```python
l = []
for i in range(len(df)):
    l.append(df["published_date"][i].strftime("%A"))
df["published_day"] = l
```

```python
df['published_month'] = df['seperated_publish_time'].dt.month
```

```python
df["published_trending_difference"] = df["published_trending_difference"].dt.days
```
